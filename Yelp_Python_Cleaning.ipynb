```python
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Projects/Yelp API/clean_yelp_food.csv')
df.head()
```

```python
# label columns
df.columns = ['Index', 'Alias', 'Name', 'Closed', 'Reviews', 'Categories', 'Rating', 'Transaction', 'Price', 'Address' ]
df.head()
```

```python
# drop closed columns
df = df.drop(columns = ['Index', 'Alias', 'Closed'])
df.head(1)
```

```python
# fixed null values in sql
df.isnull().sum()
```

```python
# fixed duplicates in sql
df.duplicated().sum()
```

```python
# clean up transaction column
df['Transaction'] = df['Transaction'].astype(str).str.replace('[\[\]]', '')
```

```python
# remove second address
df['Address'] = df['Address'].str.split(',').str[0]
```

```python
df['Price'].value_counts()
```

```python
# convert price to text
price_mapping = {'$': 'Low', '$$': 'Med', '$$$': 'High'}
df['Price'] = df['Price'].replace(price_mapping)
```
